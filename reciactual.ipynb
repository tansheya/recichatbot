```python
STOP_INGREDIENT_WORDS = {"dinner", "meal", "time", "minutes", "minute", "mins", "min", "want", "need"}

def preprocess_text(text):
    # lowercase all text for handling
    text = text.lower()

    # remove non characters
    text = re.sub(r"[^a-z0-9\s]", " ", text)

    # tokenize
    tokens = nltk.word_tokenize(text)
    return tokens

def pos_tag_tokens(tokens):
    return nltk.pos_tag(tokens)

def extract_entities(tokens, tagged):
    info = {
        "ingredients": [],
        "craving": None,
        "minutes": None,
        "diet": None,
        "wants_more": False
    }

    raw = " ".join(tokens)

    # minutes from regex
    m = re.search(r"\b(\d{1,3})\s*(?:min|mins|minute|minutes)\b", raw)
    if m:
        info["minutes"] = int(m.group(1))
    elif "quick" in tokens or "fast" in tokens:
        info["minutes"] = 15

    # cravings
    cravings = {"spicy", "healthy", "comfort", "cheap", "quick", "light", "crispy"}
    for t in tokens:
        if t in cravings:
            info["craving"] = t
            break

    # diet 
    if "vegan" in tokens:
        info["diet"] = "vegan"
    elif "vegetarian" in tokens:
        info["diet"] = "vegetarian"
    elif "halal" in tokens:
        info["diet"] = "halal"
    elif "gluten" in tokens and "free" in tokens:
        info["diet"] = "gluten-free"
    elif "pork" in tokens and "no" in tokens:
        info["diet"] = "no pork"

    # wants more recipe?
    if re.search(r"\byes\b|\banother\b|\bmore\b", raw):
        info["wants_more"] = True

    # ingredient extraction using pos tags
    for word, tag in tagged:
        if tag in ("NN", "NNS"):
            if word not in STOP_INGREDIENT_WORDS:
                info["ingredients"].append(word)

    seen = set()
    cleaned = []
    for ing in info["ingredients"]:
        if ing not in seen:
            seen.add(ing)
            cleaned.append(ing)

    info["ingredients"] = cleaned
    return info

    # remove duplicates 
    seen = set()
    cleaned = []
    for x in info["ingredients"]:
        if x not in seen:
            seen.add(x)
            cleaned.append(x)
    info["ingredients"] = cleaned

    return info
```

```python
def load_intents(path="intents.json"):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

INTENTS = load_intents()

def detect_intent(user_input):
    for intent_name, intent_data in INTENTS.items():
        for pattern in intent_data["patterns"]:
            if re.search(pattern, user_input, re.IGNORECASE):
                return intent_name
    return None

def recipe_is_allowed(recipe, memory):
    diet = memory.get("diet") or "none"
    if diet == "no pork":
        return True  
    return (diet in recipe["diet_ok"]) or ("none" in recipe["diet_ok"])

def recommend_recipe(memory):
    craving = memory.get("craving")
    ingredients = set(memory.get("ingredients", []))
    already = set(memory.get("suggested", []))
    max_mins = memory.get("minutes")

    candidates = []
    for r in RECIPE_BANK:
        if r["name"] in already:
            continue
        if not recipe_is_allowed(r, memory):
            continue
        if max_mins is not None and r["time"] > max_mins:
            continue

        score = 0
        # craving match
        if craving and craving in r["tags"]:
            score += 2
        # ingredient overlap
        score += len(ingredients.intersection(r["needs"]))

        # only keep recipes that have some chance
        if score > 0 or (max_mins is not None and r["time"] <= max_mins):
            candidates.append((score, r))

    if not candidates:
        return None

    candidates.sort(key=lambda x: x[0], reverse=True)
    # if ties, random among top few for variety
    top_score = candidates[0][0]
    top = [r for s, r in candidates if s == top_score]
    return random.choice(top)

def format_steps(steps):
    # more precise looking steps with numbering
    return "\n".join([f"{i+1}. {s}" for i, s in enumerate(steps)])

def generate_response(intent, memory):
    # suggest intent returns a recipe
    if intent in {"suggest", "craving", "time", "diet", "ingredients"}:
        recipe = recommend_recipe(memory)
        if recipe is None:
            return ("Iâ€™m out of matches ðŸ˜… Try searching using a time limit "
                    "(e.g., '15 mins'), or a craving (spicy/healthy/comfort).")

        memory["suggested"].append(recipe["name"])  
        why_bits = []
        if memory.get("craving"):
            why_bits.append(f"craving={memory['craving']}")
        if memory.get("ingredients"):
            why_bits.append(f"ingredients={', '.join(memory['ingredients'][:5])}")

        why = " / ".join(why_bits) if why_bits else "your preferences"

        return (
            f"You can make: **{recipe['name']}**\n"
            f"Time: ~{recipe['time']} mins\n"
            f"Why: matches {why}.\n"
            f"Steps:\n{format_steps(recipe['steps'])}\n"
            f"Want another suggestion? (yes/no)"
        )

    #use JSON
    templates = INTENTS[intent]["responses"]
    template = random.choice(templates)

    template = template.replace("{craving}", str(memory.get("craving", "something nice")))
    template = template.replace("{minutes}", str(memory.get("minutes", 20)))
    template = template.replace("{diet}", str(memory.get("diet", "no restrictions")))
    template = template.replace("{ingredients}", ", ".join(memory.get("ingredients", [])) or "none yet")

    return template

from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words("english"))

def build_ingredient_vocab(recipe_bank):
    """Build a set of known ingredients from RECIPE_BANK."""
    vocab = set()
    for r in recipe_bank:
        for ing in r.get("needs", []):
            vocab.add(ing.lower())
    return vocab

INGREDIENT_VOCAB = build_ingredient_vocab(RECIPE_BANK)

def extract_entities(tokens, tagged):
    """
    Extract structured info:
    - ingredients: only from known ingredient vocab (prevents 'i', 'cook', etc.)
    - craving: from a keyword list
    - minutes: from regex
    - diet: from keyword list
    """
    info = {
        "ingredients": [],
        "craving": None,
        "minutes": None,
        "diet": None
    }

    # minutes extraction 
    joined = " ".join(tokens)
    m = re.search(r"\b(\d{1,3})\s*(min|mins|minute|minutes)\b", joined)
    if m:
        info["minutes"] = m.group(1)

    # craving keywords
    craving_keywords = {"spicy", "healthy", "comfort", "cheap", "quick", "light", "crispy"}
    for t in tokens:
        if t in craving_keywords:
            info["craving"] = t

    # diet keywords
    if "gluten" in tokens and "free" in tokens:
        info["diet"] = "gluten-free"
    diet_keywords = {"vegetarian", "vegan", "halal", "no", "pork"} 
    for t in tokens:
        if t in diet_keywords:
            if t == "vegetarian":
                info["diet"] = "vegetarian"
            elif t == "vegan":
                info["diet"] = "vegan"
            elif t == "halal":
                info["diet"] = "halal"
            

    # ingredient extraction 
    clean = []
    for t in tokens:
        if t in STOPWORDS:
            continue
        if not re.fullmatch(r"[a-z]+", t):
            continue
        if len(t) < 2:
            continue
        clean.append(t)

    # accept token as ingredient only if in ingredient vocab
    for t in clean:
        if t in INGREDIENT_VOCAB:
            info["ingredients"].append(t)

    # deduplicate while preserving order
    seen = set()
    deduped = []
    for ing in info["ingredients"]:
        if ing not in seen:
            seen.add(ing)
            deduped.append(ing)
    info["ingredients"] = deduped

    return info
```

```python
import re
import json
import random
from nltk.corpus import stopwords
import nltk

STOPWORDS = set(stopwords.words("english"))

# load JSON 
with open("reci_data.json", "r", encoding="utf-8") as f:
    DATA = json.load(f)

INTENTS = DATA["intents"]
ALIASES = DATA["ingredient_aliases"]
RECIPE_BANK = DATA["recipes"]

# build ingredient vocab 
def build_ingredient_vocab(recipes, aliases):
    vocab = set()
    for r in recipes:
        for ing in r.get("needs", []):
            vocab.add(ing.lower())
    for k, v in aliases.items():
        vocab.add(k.lower())
        vocab.add(v.lower())
    return vocab

INGREDIENT_VOCAB = build_ingredient_vocab(RECIPE_BANK, ALIASES)

# helpers
def detect_intent(user_input: str):
    for intent_name, intent_data in INTENTS.items():
        for pattern in intent_data["patterns"]:
            if re.search(pattern, user_input, re.IGNORECASE):
                return intent_name
    return None

def parse_time_minutes(text: str):
    m = re.search(r"\b(\d+)\s*(min|mins|minute|minutes)\b", text.lower())
    if m:
        return int(m.group(1))
    return None

def normalize_ingredient(phrase: str):
    phrase = phrase.lower().strip()
    return ALIASES.get(phrase, phrase)

def preprocess_text(text: str):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text.split()

def extract_craving(text: str):
    cravings = {"spicy", "healthy", "comfort", "quick"}
    for c in cravings:
        if re.search(rf"\b{c}\b", text.lower()):
            return c
    return None

def extract_ingredients_from_text(text: str):
    """
    Robust ingredient extraction:
    - checks 3-word phrases then 2-word phrases (e.g., 'soy sauce', 'instant noodles')
    - checks single tokens using INGREDIENT_VOCAB and ALIASES
    """
    tokens = preprocess_text(text)
    found = set()

    junk = {
        "i","im","i'm","me","my","mine","you","your","yours",
        "want","wanna","like","feel","having","cook","make","eat",
        "dinner","meal","food","tonight","today","please",
        "mins","min","minutes","minute"
    }

    # phrase detection 
    for n in (3, 2):
        for i in range(len(tokens) - n + 1):
            phrase = " ".join(tokens[i:i+n])
            if phrase in ALIASES or phrase in INGREDIENT_VOCAB:
                found.add(normalize_ingredient(phrase))

    # single tokens
    for t in tokens:
        if t in STOPWORDS or t in junk:
            continue
        if t in INGREDIENT_VOCAB or t in ALIASES:
            found.add(normalize_ingredient(t))

    return sorted(found)

def recommend_recipe(memory):
    """
    rules:
    - avoid repeating using memory['suggested']
    - if ingredients exist must overlap at least 1 ingredient
    - if minutes exist,recipe must fit time
    """
    ingredients = set(memory["ingredients"])
    craving = memory["craving"]
    minutes = memory["minutes"]
    suggested = set(memory["suggested"])

    candidates = []

    for r in RECIPE_BANK:
        if r["name"] in suggested:
            continue

        needs = set([x.lower() for x in r.get("needs", [])])
        overlap = len(ingredients.intersection(needs))

        #overlap if user has any ingredients at all
        if len(ingredients) > 0 and overlap == 0:
            continue

        # time filter
        if minutes is not None and r.get("time_mins", 999) > minutes:
            continue

        score = 0
        score += overlap * 10
        if craving and craving in r.get("tags", []):
            score += 5

        candidates.append((score, r))

    if not candidates:
        return None

    candidates.sort(key=lambda x: x[0], reverse=True)
    return candidates[0][1]

def format_recipe_reply(recipe, memory):
    ing = ", ".join(memory["ingredients"]) if memory["ingredients"] else "none"
    craving = memory["craving"] if memory["craving"] else "none"
    minutes = memory["minutes"]

    time_line = f"~{recipe['time_mins']} mins"
    if minutes is not None:
        time_line += f" (fits your {minutes} mins)"

    steps = "\n".join([f"{i+1}. {s}" for i, s in enumerate(recipe["steps"])])

    return (
        f"You can make: **{recipe['name']}**\n"
        f"Time: {time_line}\n"
        f"Why: matches ingredients={ing} / craving={craving}\n"
        f"Steps:\n{steps}\n"
        f"Want another suggestion? (yes/no)"
    )

def fresh_memory():
    return {
        "ingredients": [],
        "craving": None,
        "minutes": None,
        "suggested": []
    }

# chatbot 
def chatbot():
    memory = fresh_memory()
    waiting_yesno = False

    print("Reci: I am Reci! Your meal assistant! Type 'exit' to exit.")
    print("Reci: Tell me ingredients, what youre craving(spicy/healthy/comfort/quick)or amount of time (e.g. 15 mins).")

    while True:
        user_input = input("You: ").strip()
        if not user_input:
            continue

        # exit
        if re.search(r"\b(bye|exit|quit|goodbye)\b", user_input, re.IGNORECASE):
            print("Reci:", random.choice(INTENTS["bye"]["responses"]))
            break

        # yes/no handling
        if waiting_yesno:
            if re.fullmatch(r"\s*yes\s*", user_input, re.IGNORECASE):
                recipe = recommend_recipe(memory)
                if not recipe:
                    print("Reci: Iâ€™m out of matches ðŸ˜… Try new inputs")
                    waiting_yesno = False
                else:
                    memory["suggested"].append(recipe["name"])
                    print("Reci:", format_recipe_reply(recipe, memory))
                    waiting_yesno = True
                continue

            if re.fullmatch(r"\s*no\s*", user_input, re.IGNORECASE):
                #resets memory when user says no
                memory = fresh_memory()
                waiting_yesno = False
                print("Reci: Ok! Tell me NEW ingredients or amount of time you have")
                continue

    
            waiting_yesno = False

        # snapshot to detect changes
        old_state = (tuple(memory["ingredients"]), memory["craving"], memory["minutes"])

        # parse time
        minutes = parse_time_minutes(user_input)
        if minutes is not None:
            memory["minutes"] = minutes

        # parse craving
        craving = extract_craving(user_input)
        if craving:
            memory["craving"] = craving

        # parse ingredients
        new_ings = extract_ingredients_from_text(user_input)
        for ing in new_ings:
            if ing not in memory["ingredients"]:
                memory["ingredients"].append(ing)

        # if user changed inputs reset suggested so it can recommend again properly
        new_state = (tuple(memory["ingredients"]), memory["craving"], memory["minutes"])
        if new_state != old_state:
            memory["suggested"] = []

        # optional intent replies
        intent = detect_intent(user_input)
        if intent == "greeting":
            print("Reci:", random.choice(INTENTS["greeting"]["responses"]))
            continue
        if intent == "suggest":
            print("Reci:", random.choice(INTENTS["suggest"]["responses"]))
            continue

        # recommend
        recipe = recommend_recipe(memory)
        if not recipe:
            print("Reci: I couldnâ€™t find a recipe that matches your inputs ðŸ˜…")
            print("Reci: Try: 'chicken garlic 15 mins spicy' or add more ingredients.")
            continue

        memory["suggested"].append(recipe["name"])
        print("Reci:", format_recipe_reply(recipe, memory))
        waiting_yesno = True

# Run
chatbot()
```
